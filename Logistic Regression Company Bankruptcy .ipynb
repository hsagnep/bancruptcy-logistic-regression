```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
warnings.filterwarnings("ignore")
```

```python
#Data Loading
df = pd.read_csv("/Users/hervinsagnep/Desktop/data.csv")
df.head()
```

```python
#Data exploration
df.isnull().sum()
df.head()
```

```python
#Looking at correlation matrix to find signs of multicollineraity
df.corr()
```

```python
#Splitting into training(70%) and testing data (30%)
train = df[:5455]
test = df[5456:]
```

```python
#Calculating VIF to remove independent variables with collinearity 
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [round(float(variance_inflation_factor(X.values, i)),2) for i in range(X.shape[1])]
    return(vif)

vif = calc_vif(train)
```

```python
#Independent Variables with a VIF less than 10 will be selected for the model
selected_variables = vif[vif.VIF < 10]
list_of_variables = [i for i in selected_variables['variables']]
```

```python
#Creating the logisitic regression model
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
x_train = train[list_of_variables[1:]]
y_train = train[list_of_variables[0]]
logisticRegr.fit(x_train, y_train)
```

```python
# Use score method to get accuracy of model
x_test = test[list_of_variables[1:]]
y_test = test[list_of_variables[0]]
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
#Prediction probabilities
predicted = logisticRegr.predict(x_test)
for i in logisticRegr.predict_proba(x_test):
    print(i)
```

```python
#Results Actual vs Predicted
data = {
    "Actual": y_test,
    "Predicted" : predicted
}
results = pd.DataFrame.from_dict(data)
results
```
